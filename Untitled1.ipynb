```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
with open('demo - Copy.html','r',encoding='utf-8') as f:
    res=f.read()
soup=BeautifulSoup(res,'html.parser')
box=soup('div',class_='tUxRFH')
len(box)
```

```python
data={'name3':[],'rom1':[],'ram':[],'color':[],'star':[],'review':[],'rear camera':[],'front camera':[],'old_price':[],'new_price':[],'discount':[],'rating':[],'display(cm)':[],'display(inch)':[]}
```

```python
data={'name3':[],'rom1':[],'ram':[],'color':[],'star':[],'review':[],'rear camera':[],'front camera':[],'old_price':[],'new_price':[],'discount':[],'rating':[],'display(cm)':[],'display(inch)':[]}
for i in box:
    data['name3'].append(i.find('div',class_='KzDlHZ').text.split('(')[0].strip())
    data['color'].append(i.find('div',class_='KzDlHZ').text.split('(')[1].split(',')[0])
    data['ram'].append(i.find('li',class_='J+igdf').text.split('|')[0])
    data['rom1'].append(i.find('li',class_='J+igdf').text.split('|')[1][0:11])
    data['star'].append(i.find('div',class_='XQDdHH').text)
    data['rating'].append(i.find('span',class_='Wphh3N').text.strip().split('Ratings')[0])
    data['review'].append(i.find('span',class_='Wphh3N').text.strip().split('Ratings')[1].strip().split('&\xa0')[1].split('Reviews')[0])
    data['rear camera'].append(i.find('li',class_='J+igdf').text.split('Display')[1].split('|')[0])
    data['front camera'].append(i.find('li',class_='J+igdf').text.split('Display')[1].split('|')[1].split(' ')[1])
    data['old_price'].append(i.find('div',class_='Nx9bqj _4b5DiR').text[1:].replace(',',""))
    data['display(cm)'].append(i.find('li',class_='J+igdf').text.split('ROM')[1].split('cm')[0])
    data['display(inch)'].append(i.find('li',class_='J+igdf').text.split('(')[1].split('inch')[0])
    try:
        data['new_price'].append(i.find('div',class_='yRaY8j ZYYwLA').text[1:].replace(',',""))
    except:
        data['new_price'].append(np.nan)
    try:
        data['discount'].append(i.find('div',class_='UkUFwK').text.split(' ')[0])
    except:
        data['discount'].append(np.nan)
        
data
```
