```python
url="https://subslikescript.com/movie/The_Golden_West-26923682"
response= requests.get(url)
response.status_code
soup=BeautifulSoup(response.text,'html.parser')
box=soup.find('article',class_='main-article')
title=box.h1.text
script=box.find('div',class_='full-script').text.strip()
with open(f"{title}.txt",'w')as f:
    f.write(script)
```

```python
url='https://subslikescript.com/movies_letter-V'
response=requests.get(url)
response.status_code
soup=BeautifulSoup(response.text,'html.parser')
box=soup.find('article',class_='main-article')
print(len(box.find_all('a')))
```

```python
links=[]
for i in box.find_all('a'):
    links.append('https://subslikescript.com'+i['href'])
links
```

```python
for i in links:
    try:
        url=i
        response=requests.get(url)
        response.status_code
        soup=BeautifulSoup(response.text,'html.parser')
        box=soup.find('article',class_='main-article')
        title=box.h1.text
        script=box.find('div',class_='full-script').text.strip()
        with open(f"{title}.txt",'w') as f:
            f.write(script)
    except:
        print('error')
```
