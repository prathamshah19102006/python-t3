```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
with  open('vishal_new.html','r',encoding='utf-8') as f:
    res=f.read()
soup=BeautifulSoup(res,'html.parser')
mobile_name=soup.find_all('div',class_="_4rR01T")
star=soup.find_all('div',class_='_3LWZlK')
ratings=soup.find_all('span',class_='_2_R_DZ')

print(mobile_name[0].text)
print(star[0].text)
print(ratings[0].text.split('&')[0].strip())
```

```python
df=pd.DataFrame({'mobile_name':[],'star':[],'ratings':[],'reviews':[],'final price':[],'old price':[],'display':[],'camera':[],'discount':[]})
```

```python
d={'mobile_name':[],'star':[],'ratings':[],'reviews':[],'final price':[],'old price':[],'display':[],'camera':[],'discount':[]}
box=soup.find_all('div',class_='_3pLy-c row')
for i in box:
    try:
        d['mobile_name'].append(i.find('div',class_="_4rR01T").text.strip())
        d['star'].append(float(i.find('div',class_='_3LWZlK').text.strip()))
        d['ratings'].append(i.find('span',class_='_2_R_DZ').text.split('&')[0][0:-8].strip())
        d['reviews'].append(i.find('span',class_='_2_R_DZ').text.split('&')[1][1:-8].strip())
        d['display'].append(i.find('ul',class_='_1xgFaf').find_all('li')[1].text.strip())
        d['camera'].append(i.find('ul',class_='_1xgFaf').find_all('li')[2].text.strip())
        d['final price'].append(i.find('div',class_='_30jeq3 _1_WHN1').text[1:].strip().replace(',',''))
        d['old price'].append(i.find('div',class_='_3I9_wc _27UcVY').text.strip())
        # d['discount'].append(i.find('div',class_='_3Ay6Sb').text.split()[0].strip()[0:i.find('div',class_='_3Ay6Sb').text.split()[0].find("%")])
        d['discount'].append(i.find('div',class_='_3Ay6Sb').text.strip()[:-5])
    except:
        d['old price'].append(np.nan)
        d['discount'].append(np.nan)
    # d['final price'].append(i.find('div',class_='_25b18c').text.strip())
df1=pd.DataFrame(d)
df1
```

```python
with open('List of state and union territory capitals in India - Wikipedia (5_28_2025 8：46：50 AM).html','r',encoding='utf-8') as f:
    res=f.read()
soup=BeautifulSoup(res,'html.parser')
box=soup.find_all('table',class_='wikitable sortable jquery-tablesorter')
box
```

```python
state=box[0]
head=state.find('thead')
col=[]
for i in head.find_all('th'):
    col.append(i.text.strip())
col[5]=col[5].split('\n')[0]
col
```

```python
body=state.find('tbody')
d={col[0]:[],col[1]:[],col[2]:[],col[3]:[],col[4]:[],col[5]:[]}
for i in body.find_all('tr'):
    l=i.find_all('td')
    d[col[0]].append(l[0].text.strip())
    d[col[1]].append(l[1].text.strip())
    d[col[2]].append(l[2].text.strip())
    d[col[3]].append(l[3].text.strip())
    d[col[4]].append(l[4].text.strip())
    d[col[5]].append(l[5].text.strip())
df=pd.DataFrame(d)
df
```
