```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
with open('Certified used Mercedes-Benz for sale.html',encoding='utf-8') as f:
    res=f.read()
soup=BeautifulSoup(res,'html.parser')
box=soup.find_all('div',class_='vehicle-details')
```

```python
data={'year':[],'name':[],'price':[],'mil':[],'rating':[]}
for i in box:
    data['year'].append(i.find('h2',class_='title').text.split(" ")[0])
    data['name'].append(i.find('h2',class_='title').text.split(" ",2)[2])
    data['mil'].append(i.find('div',class_='mileage').text.split('mi')[0].replace(",",''))
    data['price'].append(i.find('span',class_='primary-price').text[1:].replace(",",''))
    try:
        data['rating'].append(i.find('span',class_='sds-rating__count').text)
    except:
        data['rating'].append(np.nan)
```
